```
import pandas as pd
import numpy as np
x = pd.Series(np.random.normal(2,3,1000))
sr = pd.Series(['New York', 'Chicago', None, 'Lisbon', 'Rio']) 
'''*************************************
统计汇总的函数
****************************************'''
#1 Series.min() 计算序列中的最小值
print(x.min())
#2 Series.max() 计算序列中的最大值
print(x.max())
#3 Series.mean() 计算序列中的平均值
print(x.mean())
#4 Series.sum() 求和
print(x.sum())
#5 Series.count() 计数（统计非缺失元素个数）
print(sr.count())
#6 Series.size() 计数（统计所有元素个数）
print(sr.size)
#7 Series.median() 计算中位数
print(x.median())
#8 Series.var() 计算方差
print(x.var())
#9 Series.std() 计算标准差
print(x.std())
#10 Series.quantile(q=0.5) 计算任意分位数
print(x.quantile([.25, .5, .75]))
#11 Series.cov() 计算协方差
s1 = pd.Series([0.90010907, 0.13484424, 0.62036035])
s2 = pd.Series([0.12528585, 0.26962463, 0.51111198])
print(s1.cov(s2))
#12 Series.corr() 计算相差系数
print(s1.corr(s2))
#13 Series.skew() 计算偏度
print(x.skew())
#14 Series.kurt() 计算峰度
print(x.kurt())
#15 Series.mode() 计算众数
s3 = pd.Series([0.1, 0.2, 0.1])
print(s3.mode())
#16 Series.describe() 描述性统计（一次性返回多个统计结果）
print(x.describe())

#17 Series.groupby() 分组
#18 Series.aggregate() 聚合运算（可以自定义统计函数）
y = 3*x + 10 + pd.Series(np.random.normal(1,2,1000))
z = pd.Series(['A','B','C']).sample(n = 1000, replace = True)
# 重新修改z的行索引
z.index = range(1000)
# 按照z分组，统计y的组内平均值
y.groupby(by = z).aggregate(np.mean)

print("------------------------------------------------------------------------------")
a = pd.Series([1,5,10,15,25,30,3,6,8,10,5,1])
#20 Series.argmin() 查找最小值位置
print(a.argmin())
#21 Series.argmax() 查找最大值位置
print(a.argmax())
#22 Series.any()  等价于逻辑或
print(pd.Series([True, False]).any())
#23 Series.all()  等价于逻辑与
print(pd.Series([True, False]).all())
#24 Series.value_counts() 频次统计
print(a.value_counts())
#25 Series.cumsum() 运算累计和
print(a.cumsum())
#26 Series.cumprod() 运算累计积
print(a.cumprod())
#27 Series.pct_change() 运算比率（后一个元素与前一个元素的比率）  （5-1）/1
print(a.pct_change())
'''*************************************
数据清洗的函数
****************************************'''
b = pd.Series([1,5,10,15,25,30,3,6,8,10,5,1])
#28 Series.duplicated() 判断序列元素是否重复
print(b.duplicated())
#29 Series.drop_duplicates() 删除重复值
print(b.drop_duplicates())
#30 Series.hasnans()  判断序列是否存在缺失  仅返回True或False
c=pd.Series([1,2,None,5,7,0])
print(c.hasnans)
#31 Series.isnull() 判断序列元素是否为缺失 返回与序列长度一样的bool值
print(c.isnull())
#32 Series.notnull() 判断序列元素是否不为缺失 返回与序列长度一样的bool值
print(c.notnull())
#33 Series.dropna() 删除缺失值
print(c.dropna())
#34 Series.fillna() 缺失值填充
print(c.fillna(value = b.mean()))
#35 Series.ffill() 前向填充缺失值 使用缺失值的前一个元素填充
print(c.ffill())
#36 Series.bfill() 后向填充缺失值 使用缺失值的后一个元素填充
print(c.bfill())

#37 Series.dtypes() 检查数据类型
income = pd.Series(['12500元','8000元','8500元','15000元','9000元'])
print(income.dtypes)
#38 Series.astype() 强制转换数据类型
print(income.str[:-1].astype(int))
#39 pd.to_datetime() 转日期时间型
pd.to_datetime(1490195805, unit='s')
#40 Series.factorize() 因子化转换
labels, uniques = pd.factorize(['b', 'b', 'a', 'c', 'b'])
print(labels)
print(uniques)
labels, uniques = pd.factorize(['b', 'b', 'a', 'c', 'b'], sort=True)
labels, uniques = pd.factorize(['b', None, 'a', 'c', 'b'])
#41 Series.sample() 抽样
df = pd.DataFrame({'num_legs': [2, 4, 8, 0],
                   'num_wings': [2, 0, 0, 0],
                    'num_specimen_seen': [10, 2, 1, 8]},
                   index=['falcon', 'dog', 'spider', 'fish'])
df['num_legs'].sample(n=3, random_state=1)
df.sample(frac=0.5, replace=True, random_state=1)
df.sample(n=2, weights='num_specimen_seen', random_state=1)
#42 Series.where() 基于条件判断的值替换
s = pd.Series(range(5))
s.where(s > 0)
#43 Series.replace() 按值替换  不可使用正则
s = pd.Series([0, 1, 2, 3, 4])
s.replace(0, 5)
df.replace({'A': r'^ba.$'}, {'A': 'new'}, regex=True)
#44 Series.str.replace() 按值替换 可以使用正则 .str不能少
pd.Series(['foo', 'fuz', np.nan]).str.replace('f.', 'ba', regex=True)
pd.Series(['f.o', 'fuz', np.nan]).str.replace('f.', 'ba', regex=False)
#45 Series.str.split.str() 字符分割 .str不能少
house = pd.Series(['大宁金茂府 | 3室2厅 | 158.32平米 | 南 | 精装',
                   '昌里花园 | 2室2厅 | 104.73平米 | 南 | 精装',
                   '纺大小区 | 3室1厅 | 68.38平米 | 南 | 简装'])
# 取出二手房的面积，并转换为浮点型
house.str.split('|').str[2].str.strip().str[:-2].astype(float)

'''
**********************************************************************************************
                    数据筛选
**********************************************************************************************
'''
#46 Series.isin() 成员关系判断
s = pd.Series(['lama', 'cow', 'lama', 'beetle', 'lama',
               'hippo'], name='animal')
s.isin(['cow', 'lama'])

#47 Series.between() 区间判断
s = pd.Series([2, 0, 4, 8, np.nan])
s.between(1, 4)
s.between(1, 4, inclusive=False)

#48 Series.loc() 条件判断 可使用在数据框中
np.random.seed(1234)
x = pd.Series(np.random.randint(10,20,10))
# 筛选出16以上的元素
print(x.loc[x > 16])

#49 Series.iloc() 索引判断 可使用在数据框中
mydict=[[1,2,3,4],
     [100,200, 300,400],
           [1000,2000,3000,  4000 ]]
df = pd.DataFrame(mydict,columns=['a', 'b', 'c','d'])
df.iloc[[0]]
df.iloc[:3]
df.iloc[[True, False, True]]
df.iloc[lambda x: x.index % 2 == 0]
#50 Series.compress() 条件判断 
print(x.compress(x > 16))

#51 Series.nlargest() 搜寻最大的N个元素
print(x.nlargest(3))

#52 Series.nsmallest() 搜寻最小的N个元素
print(x.nsmallest(3))

#53 Series.findall() 子串查询 可使用正则
s = pd.Series(['Lion', 'Monkey', 'Rabbit'])
s.str.findall('Monkey')

'''
**********************************************************************************************
                    绘图与元素级函数
**********************************************************************************************
'''
#54 Series.hit() 绘制直方图

#55 Series.plot() 可基于kind的参数绘制更多的图形 可绘制饼图、箱线图、折线图等

#56 Series.map() 元素映射

#57 Series.apply() 基于自定义函数进行元素级的操作

np.random.seed(123)
import matplotlib.pyplot as plt
x = pd.Series(np.random.normal(10,3,1000))
# 绘制x直方图
x.hist()
# 显示图形
plt.show()

# 绘制x的箱线图
x.plot(kind='box')
plt.show()

installs = pd.Series(['1280万','6.7亿','2488万','1892万','9877','9877万','1.2亿'])
# 将安装量统一更改为“万”的单位
def transform(x):
    if x.find('亿') != -1:
        res = float(x[:-1])*10000
    elif x.find('万') != -1:
        res = float(x[:-1])
    else:
        res = float(x)/10000
    return res
print(installs.apply(transform))

frame = pd.DataFrame({'item':['ball', 'mug', 'pen', 'pencil', 'ashtray'],
                      'price':[1, 2, 3, 4, 5]})

frame.price = frame.price.map(lambda x: x + 1)
print(frame)
'''
**********************************************************************************************
                    时间序列函数
**********************************************************************************************
'''
dfdate=pd.to_datetime(
    pd.Series(['2019-10-1 13:01:10','2019-10-2 13:02:11','2019-10-3 13:03:12','2019-10-4 13:04:15','2019-10-5 13:05:16','2019-10-6 13:06:17','2019-10-31 13:08:19']))
#58 Series.dt.date 抽取出日期值 Series为日期时间型序列，年-月-日
print(dfdate.dt.date)
#59 Series.dt.time 抽取出时间
print(dfdate.dt.time)
#60 Series.dt.year 抽取出年
print(dfdate.dt.year)
#61 Series.dt.month 抽取出月
print(dfdate.dt.month)
#62 Series.dt.day 抽取出日
print(dfdate.dt.day)
#63 Series.dt.hour 抽取出时
print(dfdate.dt.hour)
#64 Series.dt.minute 抽取出分钟
print(dfdate.dt.minute)
#65 Series.dt.second 抽取出秒
print(dfdate.dt.second)
#66 Series.dt.quarter 抽取出季度
print(dfdate.dt.quarter)
#67 Series.dt.weekday 抽取出星期几 返回数值型
print(dfdate.dt.weekday)
#68 Series.dt.weekday_name 抽取出星期几 返回字符型
print(dfdate.dt.weekday_name)
#69 Series.dt.week 抽取出年中的第几周
print(dfdate.dt.week)
#70 Series.dt.dayofyear 抽取出年中的第几天
print(dfdate.dt.dayofyear)
#71 Series.dt.daysinmonth 抽取出月对应的最大天数
print(dfdate.dt.daysinmonth)
#72 Series.dt.is_month_start 判断日期是否为当月的第一天
print(dfdate.dt.is_month_start)
#73 Series.dt.is_month_end 判断日期是否为当月的最后一天
print(dfdate.dt.is_month_end)
#74 Series.dt.is_quarter_start 判断日期是否为当季度的第一天
print(dfdate.dt.is_quarter_start)
#75 Series.dt.is_quarter_end 判断日期是否为当季度的最后一天
print(dfdate.dt.is_quarter_end)
#76 Series.dt.is_year_start 判断日期是否为当年的第一天
print(dfdate.dt.is_year_start)
#77 Series.dt.is_year_end 判断日期是否为当年的最后一天
print(dfdate.dt.is_year_end)
#78 Series.dt.is_leap_year 判断日期是否为闰年
print(dfdate.dt.is_leap_year)
'''
**********************************************************************************************
                    其他函数
**********************************************************************************************
'''
#79 Series.append 序列元素的追加 需要指定其他的序列
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
s3 = pd.Series([4, 5, 6],index=[3,4,5])
s1.append(s2)
s1.append(s3)
s1.append(s2,ignore_index=True)
#s1.append(s2,verify_integrity=True) #如果Index一样会报错
#80 Series.diff 一阶差分
s = pd.Series([1, 1, 2, 3, 5, 8])
s.diff()
s.diff(periods=3)
s.diff(periods=-1)
#81 Series.round 元素的4舍5入
s = pd.Series([1.1, 1.5, 2.6, 3.2, 5.8, 8.8])
s.round()
#82 Series.sort_values 按值排序
s = pd.Series([np.nan, 1, 3, 10, 5])
s.sort_values(ascending=True)
s.sort_values(ascending=False)
#83 Series.sort_index 按索引排序
s = pd.Series(['a', 'b', 'c', 'd'], index=[3, 2, 1, 4])
s.sort_index()
s.sort_index(ascending=False)
s.sort_index(inplace=True)
s

s = pd.Series(['a', 'b', 'c', 'd'], index=[3, 2, 1, np.nan])
s.sort_index(na_position='first')

arrays = [np.array(['qux', 'qux', 'foo', 'foo',
                     'baz', 'baz', 'bar', 'bar']),
           np.array(['two', 'one', 'two', 'one',
                     'two', 'one', 'two', 'one'])]
s = pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=arrays)
s.sort_index(level=1)
s.sort_index(level=1, sort_remaining=False)
#84 Series.to_dict 转为字典
s = pd.Series([1, 2, 3, 4])
s.to_dict()
#85 Series.tolist 转为列表
s = pd.Series([1, 1, 2, 3, 5, 8,8,5,6])
s.tolist()
#86 Series.unique 元素排重
s = pd.Series([1, 1, 2, 3, 5, 8,8,5,6])
s.unique()
```
